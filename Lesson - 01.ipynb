# www.boi.org.il

```python
from lxml import html
import requests
 
url = "https://www.boi.org.il/currency.xml"

response = requests.get(url)
tree = html.fromstring(response.content)

d2s_rate = float(tree[1][4].text)

print ('1 USD =', d2s_rate, 'NIS' )
```

1. # Lesson 2: List, Tuples, Set, Dictionary, Loops, and conditions


### 1.  List is a collection which is ordered and changeable. Allows duplicate members.
### 2.  Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
### 3.  Set is a collection which is unordered and unindexed. No duplicate members.
### 4.  Dictionary is a collection which is unordered, changeable and indexed. No duplicate members.

```python
a = [1, 2, 8, 4, 1]
b = (1, 2, 8, 4, 1)
c = {1, 2, 8, 4, 1}

print (type(a))
print (type(b))
print (type(c))
```

```python
print ('len(a) is', len(a))
print ('len(b) is', len(b))
print ('len(c) is', len(c))
```
